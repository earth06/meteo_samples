```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.gridspec import GridSpec
plt.style.use("ggplot")
```

```python
df1=pd.read_html("https://www.data.jma.go.jp/obd/stats/etrn/view/10min_s1.php?prec_no=51&block_no=47636&year=2023&month=3&day=12&view=")[0]
```

```python
df2=pd.read_html("https://www.data.jma.go.jp/obd/stats/etrn/view/10min_s1.php?prec_no=51&block_no=47636&year=2023&month=3&day=13&view=")[0]
```

```python
columns=["time","ps","p0","precip","temp", "rh","wind_mean","wind_dir_mean","wind_max","wind_dir_max","day_time"]
df1.columns=columns
df2.columns=columns
```

```python
dfall=pd.concat([df1,df2])
dfall.index=pd.date_range("2023-03-12 00:10","2023-03-14 00:00", freq="10T")
dfall
```

```python
dfall["precip"]=dfall["precip"].replace("--",0).astype(float)
```

```python
directions=["北","北北東","北東","東北東","東","東南東","南東","南南東","南","南南西","南西","西南西","西","西北西","北西","北北西"]

winddirection_to_theata={
    direction:i*22.5 for i,direction in enumerate(directions)
}
winddirection_to_theata["静穏"]=np.nan

def replace(x):
    if x is np.nan:
        return np.nan
    if x=='×':
        return np.nan
    return winddirection_to_theata[x]

dfall["wind_dir_mean"]=dfall["wind_dir_mean"].apply(replace)
```

```python
dfall["u_wind"]=dfall["wind_mean"]*np.sin(np.deg2rad(dfall["wind_dir_mean"]+180))
dfall["v_wind"]=dfall["wind_mean"]*np.cos(np.deg2rad(dfall["wind_dir_mean"]+180))
```

```python
dfall["Y"]=0.5
```

```python
dfall["date_time"]=dfall.index
```

```python
dfall_sel=dfall.loc["2023-03-01 22:00":"2023-03-02 07:00"]
```

```python
begin=pd.to_datetime("2023-03-13 3:00")
end=pd.to_datetime("2023-03-13 10:00")
dfall_sel=dfall.loc[begin:end].copy()


fig=plt.figure(layout="constrained",figsize=(16,8))
gs=GridSpec(5,1)
ax1=fig.add_subplot(gs[0,:])
ax1.tick_params(labelbottom=False,bottom=False,labelleft=False, left=False)
ax2=fig.add_subplot(gs[1:4,:])
ax2.tick_params(labelbottom=False,bottom=False)
ax2.yaxis.set_major_locator(plt.MultipleLocator(2))
ax2.yaxis.set_minor_locator(plt.MultipleLocator(1))
ax2_2=ax2.twinx()
ax2_2.yaxis.set_major_locator(plt.MultipleLocator(2))
ax2_2.yaxis.set_minor_locator(plt.MultipleLocator(1))
ax2.grid(which="both",axis="both",linestyle="--")
ax2_2.grid(visible=False)

ax3=fig.add_subplot(gs[4,:])
diff=16
t_min=6
p_min=1006

#風速
ax1.barbs(dfall_sel["date_time"], dfall_sel["Y"], dfall_sel["u_wind"], dfall_sel["v_wind"],barb_increments={"half":1,"full":2,"flag":10})
ax1.grid(which="both",axis="both",linestyle="--")
ax1.set_ylabel("wind[m/s]")
ax1.set_ylim(0.48,0.52)

#気温・気圧
p1=ax2.plot(dfall_sel["date_time"], dfall_sel["temp"],c="C0",label="temp",marker=".")
p2=ax2_2.plot(dfall_sel["date_time"], dfall_sel["p0"], c="C1", label="p0",marker=".")
ax2.set_ylim(t_min,t_min+diff)
ax2_2.set_ylim(p_min,p_min+diff)
ax2.legend(handles=[p1[0],p2[0]])
ax2.set_ylabel("temperature[℃]")
ax2_2.set_ylabel("sea level pressure[hPa]")


#降水量
#dfall_sel["precip"].plot(kind="bar",ax=ax3)
ax3.bar(dfall_sel["date_time"], dfall_sel["precip"],width=0.002)
ax3.grid(which="both",axis="both",linestyle="--")
ax3.set_ylabel("precip[mm]")
# ax3.set_ylim(0,2)


ax1.set_xlim(begin,end)
ax2.set_xlim(begin,end)
ax3.set_xlim(begin,end)
```
