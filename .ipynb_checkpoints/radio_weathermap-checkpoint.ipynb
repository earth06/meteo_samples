```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import xarray as xr
import metpy.calc as mcalc
from metpy.units import units
```

```python
df=pd.read_html("https://hattoritaka.web.fc2.com/MYORG/itiran.html")[0]
df2=df.iloc[2:].copy()
```

```python
df2.columns=["idx","st_name","st_name_kana","lat_deg","lat_min","lon_deg","lon_min","alt","attr"]
df2.replace("-",np.nan,inplace=True)
df2=df2.astype({"lat_deg":"float","lat_min":"float", "lon_deg":"float","lon_min":"float"})
df2["lat"]=df2["lat_deg"]+df2["lat_min"]/60.0
df2["lon"]=df2["lon_deg"]+df2["lon_min"]/60.0
dfout=df2[["st_name","st_name_kana","lat","lon","alt"]]
dfout.to_csv("radio_st_list.csv",index=None)
```

```python
st_master=pd.read_csv("./radio_st_list.csv")
st_master
```

```python
mlon=plt.MultipleLocator(10)
mlat=plt.MultipleLocator(10)
```

```python
fig=plt.figure(figsize=(20,12))
ax=fig.add_subplot(1,1,1,projection=ccrs.LambertConformal(central_longitude=140, central_latitude=30))
ax.coastlines(resolution="50m")
gl_minor=ax.gridlines(draw_labels=False, xlocs=plt.MultipleLocator(2), ylocs=plt.MultipleLocator(2),alpha=0.5)
gl=ax.gridlines(draw_labels=True,xlocs=mlon,ylocs=mlat, linewidth=2,alpha=0.5)
ax.set_extent([110,164,14,52],crs=ccrs.PlateCarree())

for lon,lat in zip(st_master["lon"], st_master["lat"]):
    ax.plot(lon,lat, marker="o", transform=ccrs.PlateCarree(), c="w", markeredgecolor="k",markersize=10)
```
